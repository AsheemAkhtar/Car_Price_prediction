## Creating Labels for Each Categorical Column

```python
def LABEL_ENCODING(c1):
    from sklearn import preprocessing
    label_encoder = preprocessing.LabelEncoder()
    df_train[c1]= label_encoder.fit_transform(df_train[c1])
    df_train[c1].unique()
```
